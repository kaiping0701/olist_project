```python
df = pd.DataFrame({
    "name": ["A", "B", "C", "D", "E"],
    "age": [20, 30, 25, 40, 35],
    "gender": ["M", "F", "M", "F", "M"],
    "score": [88, 92, 75, 60, 95]
})

df
```

```python
#第 1 題：取 index = 2 的整列 row
df.loc[2]
```

```python
#第 2 題：取 index = 4 的 name 欄位值
df.loc[4,"name"]
```

```python
#第 3 題：取 index = 0 到 2（含 2）的所有欄位
df.loc[0:2,:]
```

```python
# 第 4 題：取 index = 1 到 3 的 name 與 age 欄位
df.loc[1:3,["name","age"]]
```

```python
#第 5 題：用 .loc 取得所有男性（gender = "M"）的 row

df.loc[df["gender"]=="M",:]

```

```python
#第 6 題：取 index = 1 的 row 但只要 score 欄位
print(df.loc[1,"score"])
```

```python
# 第 7 題：一次取兩個不連續的 row（index 0 和 3）
df.loc[[0,3],:]
```

```python
#第 8 題：取所有 row，但只要 columns = ["name", "score"]
df.loc[:,["name","score"]]
```

```python
#第 9 題（進階）：取 score >= 80 的所有人的 name 與 score

df.loc[df["score"]>=80,["name","score"]]
```

```python
#第 10 題（進階）：取 gender = F 的人，但只要 age 與 score

df.loc[df["gender"]=="F",["age","score"]]
```

```python
df = pd.DataFrame({
    "name": ["A", "B", "C", "D", "E"],
    "age": [20, 30, 25, 40, 35],
    "gender": ["M", "F", "M", "F", "M"],
    "score": [88, 92, 75, 60, 95]
})

df
```

```python
#第 1 題：取第 0 列（第一筆資料）

df.iloc[0]
```

```python
#第 2 題：取第 2 列（第三筆）的第 1 欄（age）
print(df.iloc[2,1])
```

```python
#第 3 題：取第 1 到第 3 列（不含 3）
df.iloc[1:3]
```

```python
#第 4 題：取第 0 到第 2 列（含 2），但只取第 0 和第 3 欄（name、score）
df.iloc[0:3,[0,3]]
```

```python
#第 5 題：取第 4 列（最後一列）的最後一欄（score）
print(df.iloc[4,-1])
```

```python
#第 6 題：一次取不連續 row（第 0、2、4 列）
df.iloc[[0,2,4]]
```

```python
#第 7 題：一次取不連續 columns（只取 age 與 score）
df.iloc[:,[1,3]]
```

```python
#第 8 題：取前三列（0,1,2）的所有欄位的所有值
df.iloc[[0,1,2]]
```

```python
#第 9 題：取第 1 到第 4 列（不含 4）與第 1 到第 3 欄（不含 3）

df.iloc[1:4,1:3]
```

```python
#第 10 題（進階）：取第 0、1、4 列，並且只取 name 與 score 兩欄

df.iloc[[0,1,4],[0,3]]
```
