```python
import pandas as pd
import numpy as np
```

```python


df = pd.DataFrame({
    "order_id": [
        "O001", "O002", "O003", "O004", "O005",
        "O006", "O007", "O008", "O009", None
    ],
    "customer_name": [
        "Alice", "Bob", "Carol", "David", "Emily",
        "Frank", None, "Helen", "Ian", "Jane"
    ],
    "age": [
        25, 32, None, 28, 40,
        None, 22, 35, 30, None
    ],
    "order_date": [
        "2024-03-01", "2024-03-02", "2024/03/05", "March 10, 2024", "2024-03-15",
        None, "2024-03-20", "2024/03/21", "2024-03-22", "March 25, 2024"
    ],
    "price": [
        "100", "150.5", None, "250.5", "300",
        "abc", "80", "120", None, "500"
    ],
    "status": [
        "Completed", "completed", "Completed", "CANCEL", "cancel",
        None, "Completed", "completed", "Completed", "CANCEL"
    ],
    "review": [
        None, "Good product", "Excellent", None, "Bad",
        "Okay", None, "Very good", "Excellent", None
    ]
})

df
```

```python
df["review"].isna()
```

```python
df["review"].isnull()
```

```python
df["review"].fillna("我是補充值")
```

```python
df["age"].astype("object")
```

```python
df.loc[:,["age","review"]].fillna({"age":"未知年齡","review":"未知評價"})
# 針對某些欄位補充
```

```python
df["age"].astype("Int64")
```

```python
df["age"].fillna(df["age"].median()) # 用中位數補空值
```

```python
df["age"].fillna(df["age"].mean()) # 用平均數補充值
```

```python
df["age"].fillna(method="ffill") # 用前一個有效值填充
```

```python
df["age"].fillna(method="bfill") # 用後一個有效值填充
```

```python
df
```

```python
df.dropna() # 默認axis=0 ， 有na row全部丟掉
```

```python
df.dropna(axis=1) # 刪掉有na 的 column
```

```python
df.dropna(axis=0,how="any") # 默認any 有缺失就刪除
```

```python
df.dropna(axis=0,how="all") # 一個row的資料全部都是na才刪除
```

```python
# 第 1 題：刪除任何欄位有缺失值的 row
df.dropna(axis=0)
```

```python
#第 2 題：刪除 order_id 缺失的 row（因為 order_id 是主鍵）
df.dropna(axis=0,subset=["order_id"])
```

```python
#第 3 題：刪除缺失欄位超過 3 個的 row
df.dropna(axis=0,thresh=4) # 保留非缺失至少3個
```

```python
#第 4 題：只刪除「全為缺失」的 row
df.dropna(axis=0,how="all")
```

```python
#第 5 題：刪除有缺失的 columns
df.dropna(axis=1,how="any")
```

```python
#第 6 題：保留「至少 5 個非缺失值」的 row (缺失三個就刪掉)
df.dropna(axis=0,thresh=5)
```

```python
#第 7 題：刪除只有 review 缺失的 row（其他欄位不管）
df.dropna(axis=0,subset="review")
```

```python
#第 8 題：萃取出 age 有缺失的 row（不是 drop，是篩選）
df[df["age"].isna()]
```

```python
thresh = int(len(df) * 0.5)   # 至少要這麼多非缺失才保留
df_drop = df.dropna(axis=1, thresh=thresh)
```
